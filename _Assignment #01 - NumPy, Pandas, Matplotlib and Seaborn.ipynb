```python
# 1.Read the dataset to python environment.
import pandas as pd

dataframe = pd.read_excel('iris.xls')
 
print(dataframe)
```

```python
# 2.Display the columns in the dataset.
dataframe.columns
```

```python
# 3.Calculate the mean of each column of the dataset.
mean1 = dataframe['SL'].mean()
print('SL:mean =',mean1)
mean2 = dataframe['SW'].mean()
print('SW:mean =',mean2)
mean3 = dataframe['PL'].mean()
print('PL:mean =',mean3)
mean4 = dataframe['PW'].mean()
print('PW:mean =',mean4)
```
