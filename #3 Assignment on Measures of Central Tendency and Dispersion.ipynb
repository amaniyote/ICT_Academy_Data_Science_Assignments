```python
#3 Who scores the most on average for math, reading and writing based on
#3.1 Gender
dataset1 = dataset[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').mean()
dataset1
```

```python
#Conclusion: Female students have higher average reading and writing while male students have higher average in maths.
```

```python
#3.2 Test preparation course
dataset2= dataset[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').mean()
dataset2
```

```python
#Conclusion: Students who did test preparation course have higher average score across all categories
```

```python
#4 What do you think about the scoring variation for math, reading and writing based on
#4.1 gender
gen= dataset[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').std()
gen
```

```python
#Conclusion: Male students have less variations across each of the scores
```

```python
#4.2 Test preparation course
test= dataset[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').std()
test
```
